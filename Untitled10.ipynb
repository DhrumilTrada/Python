# Reduce fraction

```python
num = int(input("Enter numerator "))
de = int(input("Enter denominator "))
if(de==0):
    print("ZERO DIVISION ERROR")
else: 
    if(num%de == 0):
        print("REDUCED TO--> ",num//de)
    else:
        newnum = 0
        newde = 0
        for i in range(num,1,-1):
            if((de%i == 0) & (num%i == 0)):
                num = num//i
                de = de//i
        print("REDUCED TO--> ",num,"/",de)
```
